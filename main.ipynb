```python
import pandas as pd
import matplotlib.pyplot as plt

# --- Step 1: Read the data file ---
# The data files have comments at the top starting with '#'.
# We can tell pandas to skip these rows.
# Let's start with the electron-recoil events for TUM40.
file_path = 'D:\Programming\Python\Cresst\Files\TUM40_eRecoils.dat'
df_tum40_e = pd.read_csv(file_path, comment='#', header=None, names=['Energy [keV]'])

# --- Step 2: Read the cut-survival probability file ---
# Let's get the probability for TUM40 electron recoils.
file_path_eff = 'D:\Programming\Python\Cresst\Files\TUM40_eff_eRecoils_e.dat'
df_tum40_eff_e = pd.read_csv(file_path_eff, comment='#', header=None, names=['Energy [keV]', 'Cut-Survival Probability'])

# --- Step 3: Inspect the dataframes ---
print("TUM40 Electron Recoils Data:")
print(df_tum40_e.head())
print("\n")
print("TUM40 Electron Recoils Cut-Survival Probability:")
print(df_tum40_eff_e.head())

# --- Step 4: Visualize the data ---
# Let's plot the energy spectrum, similar to Figure 3 in the document.
plt.figure(figsize=(10, 6))
plt.hist(df_tum40_e['Energy [keV]'], bins=100, color='blue', histtype='step', density=False)
plt.title('Energy Spectrum for TUM40 Electron-Recoil Events')
plt.xlabel('Energy [keV]')
plt.ylabel('Counts per bin')
plt.yscale('log') # The document's plots are semi-log, but you can choose.
plt.show()

# Let's plot the cut-survival probability, similar to Figure 4.
plt.figure(figsize=(10, 6))
plt.plot(df_tum40_eff_e['Energy [keV]'], df_tum40_eff_e['Cut-Survival Probability'], color='black', label='e recoils')
plt.title('Cut-Survival Probability for TUM40')
plt.xlabel('Energy [keV]')
plt.ylabel('Cut-survival probability $\\epsilon$')
plt.grid(True)
plt.show()

# --- Step 5: Explore the acceptance region data ---
# You can repeat the steps above for the other files.
# For example, let's plot the energy spectrum for the TUM40 dark matter acceptance region.
file_path_ar = 'D:\Programming\Python\Cresst\Files\TUM40_AR.dat'
df_tum40_ar = pd.read_csv(file_path_ar, comment='#', header=None, names=['Energy [keV]'])

plt.figure(figsize=(10, 6))
plt.hist(df_tum40_ar['Energy [keV]'], bins=100, color='red', histtype='step', density=False)
plt.title('Energy Spectrum for TUM40 Acceptance Region')
plt.xlabel('Energy [keV]')
plt.ylabel('Counts per bin')
plt.show()
```

 Figure 3: Energy Spectra for Electron-Recoil Events
This figure shows the energy distribution of events in the electron-recoil band for both TUM40 and Lise.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data for TUM40 and Lise electron recoils
df_tum40_e = pd.read_csv('D:\Programming\Python\Cresst\Files\TUM40_eRecoils.dat', comment='#', header=None, names=['Energy [keV]'])
df_lise_e = pd.read_csv('D:\Programming\Python\Cresst\Files\Lise_eRecoils.dat', comment='#', header=None, names=['Energy [keV]'])

# Create a figure with two subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6), sharey=True)

# Plot for TUM40
ax1.hist(df_tum40_e['Energy [keV]'], bins=1000, color='blue', histtype='stepfilled')
ax1.set_title('TUM40')
ax1.set_xlabel('energy E [keV]')
ax1.set_ylabel('counts [keV$^{-1}$ kg$^{-1}$ day$^{-1}$]')
ax1.set_xlim(0, 100)
ax1.set_ylim(0, 25)
ax1.grid(True, which='both', linestyle='--')

# Plot for Lise
ax2.hist(df_lise_e['Energy [keV]'], bins=10000, color='blue', histtype='stepfilled')
ax2.set_title('Lise')
ax2.set_xlabel('energy E [keV]')
ax2.set_ylabel('counts [keV$^{-1}$ kg$^{-1}$ day$^{-1}$]')
ax2.set_xlim(0, 100)
ax2.set_ylim(0, 25)
ax2.grid(True, which='both', linestyle='--')

# Adjust layout and show the plot
plt.suptitle('Figure 3: Energy spectra for electron-recoil events for TUM40 (left) and Lise(right).', fontsize=14)
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```

Figure 4: Cut-Survival Probabilities (Electron-Recoil Band)
This figure shows the cut-survival probabilities for different event types within the electron-recoil band for TUM40 and Lise.


```python
import pandas as pd
import matplotlib.pyplot as plt

# Update these file paths to match the exact location of your files.
# For example, if they are in a 'Files' subdirectory:
# 'e recoils': 'Files/TUM40_eff_eRecoils_e.dat'
files_tum40 = {
    'e recoils': r'D:\Programming\Python\Cresst\Files\TUM40_eff_eRecoils_e.dat',
    'O recoils': r'D:\Programming\Python\Cresst\Files\TUM40_eff_eRecoils_O.dat',
    'Ca recoils': r'D:\Programming\Python\Cresst\Files\TUM40_eff_eRecoils_Ca.dat',
    'W recoils': r'D:\Programming\Python\Cresst\Files\TUM40_eff_eRecoils_W.dat'
}

files_lise = {
    'e recoils': r'D:\Programming\Python\Cresst\Files\Lise_eff_eRecoils_e.dat',
    'O recoils': r'D:\Programming\Python\Cresst\Files\Lise_eff_eRecoils_O.dat',
    'Ca recoils': r'D:\Programming\Python\Cresst\Files\Lise_eff_eRecoils_Ca.dat',
    'W recoils': r'D:\Programming\Python\Cresst\Files\Lise_eff_eRecoils_W.dat'
}

# Create a figure with two subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6), sharey=True)

# Plot for TUM40
ax1.set_title('TUM40')
ax1.set_xlabel('energy [keV]')
ax1.set_ylabel('cut-survival probability $\\epsilon$')
ax1.set_xlim(0, 10)
ax1.set_ylim(0, 1.0)
ax1.grid(True, which='both', linestyle='--')

# Plot the lines for TUM40
colors = {'e recoils': 'black', 'O recoils': 'red', 'Ca recoils': 'blue', 'W recoils': 'green'}
for label, file in files_tum40.items():
    df = pd.read_csv(file, comment='#', header=None, names=['Energy [keV]', 'Probability'], delimiter='\t')
    ax1.plot(df['Energy [keV]'], df['Probability'], color=colors[label], label=label)
ax1.legend()

# Plot for Lise
ax2.set_title('Lise')
ax2.set_xlabel('energy [keV]')
ax2.set_ylabel('cut-survival probability $\\epsilon$')
ax2.set_xlim(0, 10)
ax2.set_ylim(0, 1.0)
ax2.grid(True, which='both', linestyle='--')

# Plot the lines for Lise
for label, file in files_lise.items():
    df = pd.read_csv(file, comment='#', header=None, names=['Energy [keV]', 'Probability'], delimiter='\t')
    ax2.plot(df['Energy [keV]'], df['Probability'], color=colors[label], label=label)
ax2.legend()

# Adjust layout and show the plot
plt.suptitle('Figure 4: Cut-survival probabilities for TUM40 (left) and Lise (right) for electron and nuclear-recoil events within the electron-recoil band.', fontsize=14)
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the test case data for TUM40 and Lise
df_tum40_test = pd.read_csv('D:/Programming/Python/Cresst/Files/TUM40_TestCase.dat', comment='#', header=None, names=['Mass [GeV/c^2]', 'Cross Section [cm^2]'])
df_lise_test = pd.read_csv('D:/Programming/Python/Cresst/Files/Lise_TestCase.dat', comment='#', header=None, names=['Mass [GeV/c^2]', 'Cross Section [cm^2]'])

# Create a figure and plot
plt.figure(figsize=(10, 8))

# Plot for TUM40
plt.plot(df_tum40_test['Mass [GeV/c^2]'], df_tum40_test['Cross Section [cm^2]'], color='red', label='TUM40')

# Plot for Lise
plt.plot(df_lise_test['Mass [GeV/c^2]'], df_lise_test['Cross Section [cm^2]'], color='blue', label='Lise')

# Set the axis to log scale
plt.xscale('log')
plt.yscale('log')

# Set labels and title
plt.title('Figure 7: Upper 90% exclusion limits for the cross section of spin-independent elastic scattering of dark-matter particles of nuclei for TUM40 and Lise.')
plt.xlabel('dark matter particle mass [GeV/c$^2$]')
plt.ylabel('dark matter particle - nucleon cross section [cm$^2$]')
plt.grid(True, which="both", linestyle='--')
plt.legend()

# Adjust plot limits to match the original figure
plt.xlim(0.8, 40)
plt.ylim(10**-43, 10**-36)

plt.tight_layout()
plt.show()
```
